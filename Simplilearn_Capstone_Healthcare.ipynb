## Week 3

### Data Modeling

```python
# Here we will be using 3 models for classification.
# The algorithms that would be used here are 
# 1) Knn
# 2) Logistic Regression
# 3) Random Forest 
# 4) SVC
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import metrics
```

```python
#Lets assign the features to the X variable and the response to the y variable 
X = df.iloc[:,0:8]
y = df.Outcome
```

```python
#Train Test Split
test_size = 0.30 # taking 70:30 training and test set
seed =1 # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
from sklearn import preprocessing

X_train = preprocessing.scale(X_train)
X_test = preprocessing.scale(X_test)
```

### KNN 

```python
KNN_model = KNeighborsClassifier(n_neighbors=3)
KNN_model.fit(X_train, y_train)
```

```python
KNN_predict = KNN_model.predict(X_test)
KNN_score = KNN_model.score(X_test,y_test)
#print("KNN Classifier Prediction :",KNN_predict)
print("KNN Classifier Score :",KNN_score)
```

```python
print("Accuracy Score of KNN Model:")
print(metrics.accuracy_score(y_test,KNN_predict))
print("\n","Classification Report:")
print(metrics.classification_report(y_test,KNN_predict),'\n')
```

### Logistic Regression

```python
#Logistic Regression
logReg = LogisticRegression()
logReg.fit(X_train,y_train)
```

```python
logReg_predict = logReg.predict(X_test)
logReg_score = logReg.score(X_test,y_test)
# print("Logistic Regression Prediction :",logReg_predict)
print("Logistic Regression Score :",logReg_score)
```

```python
print("Accuracy Score of Logistic Regression Model:")
print(metrics.accuracy_score(y_test,logReg_predict))
print("\n","Classification Report:")
print(metrics.classification_report(y_test,logReg_predict),'\n')
```

### Random Forest

```python
Random_forest = RandomForestClassifier(n_estimators=5)
Random_forest.fit(X_train,y_train)
```

```python
RF_predict = Random_forest.predict(X_test)
RF_score = Random_forest.score(X_test,y_test)
# print("Random Forest Prediction :",RF_predict)
print("Random Forest Score :",RF_score)
```

```python
print("Accuracy Score of Random Forest Classifier Model:")
print(metrics.accuracy_score(y_test,RF_predict))
print("\n","Classification Report:")
print(metrics.classification_report(y_test,RF_predict),'\n')
```

### Support Vector Classifier (SVC)

```python
SVC_model = SVC(probability=True)
SVC_model.fit(X_train,y_train)
```

```python
SVC_predict = SVC_model.predict(X_test)
SVC_score = SVC_model.score(X_test,y_test)
# print("Support Vector Classifier Prediction :",SVC_predict)
print("Support Vector CLassifier Score :",SVC_score)
```

```python
print("Accuracy Score of Support Vector Classifier SVC Model:")
print(metrics.accuracy_score(y_test,SVC_predict))
print("\n","Classification Report:")
print(metrics.classification_report(y_test,SVC_predict),'\n')
```

## Week 4

```python
# Classification report - Analyzing Sensitivity, Specificity, AUC (ROC curve), etc
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#KNN
print("\n","KNN Classifier ROC Curve")
knn_prob=KNN_model.predict_proba(X_test)
knn_prob1=knn_prob[:,1]
fpr,tpr,thresh=metrics.roc_curve(y_test,knn_prob1)
roc_auc_knn=metrics.auc(fpr,tpr)
print('AUC Value: ',roc_auc_knn)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.3f'%roc_auc_knn)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```

```python
# Logistic Regresssion
print("\n","Logistic Regression ROC Curve")
logReg_prob=logReg.predict_proba(X_test)
logReg_prob1=logReg_prob[:,1]
fpr1,tpr1,thresh1=metrics.roc_curve(y_test,logReg_prob1)
roc_auc_logReg=metrics.auc(fpr1,tpr1)
print('AUC Value: ',roc_auc_logReg)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr1,tpr1,'b',label='AUC Score = %0.3f'%roc_auc_logReg)
plt.plot(fpr1,fpr1,'r--',color='red')
plt.legend()
```

```python
#Random Forest Classifier
print("\n","Random Forest Classifier ROC Curve")
RF_prob=Random_forest.predict_proba(X_test)
RF_prob1=logReg_prob[:,1]
fpr2,tpr2,thresh2=metrics.roc_curve(y_test,RF_prob1)
roc_auc_RF=metrics.auc(fpr2,tpr2)
print('AUC Value: ',roc_auc_RF)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.3f'%roc_auc_RF)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```

```python
#SVC
print("\n","SVC Classifier ROC Curve")
svc_prob=SVC_model.predict_proba(X_test)
svc_prob1=svc_prob[:,1]
fpr4,tpr4,thresh4=metrics.roc_curve(y_test,svc_prob1)
roc_auc_svc=metrics.auc(fpr4,tpr4)
print('AUC Value: ',roc_auc_svc)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.3f'%roc_auc_svc)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```
